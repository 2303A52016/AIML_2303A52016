Part 2 â€“ Visualization of Data using matplotlib, pyplots Packages[CO2]


Let's suppose we collected age in 'years' and height in 'cm' and weight in 'kg' information for patients visiting a clinic

```python
people = ['kiran', 'arun', 'vijay', 'varun', 'ravi']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 173, 159]
weight=[45, 55, 65, 75, 77]
```

```python
import matplotlib.pyplot as plt

plt.scatter(weight, height)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight vs Height')
plt.show()
```

Now, Representing the given data in Bar Graph

```python

plt.bar(x=people, height=age, color = 'blue', alpha = 0.6)
plt.xticks(fontsize = 16, rotation = 45) # Set fontsize for x-axis tick labels
plt.yticks(fontsize = 16) # Set fontsize for y-axis tick labels
plt.xlabel('People')
plt.ylabel('Age')
plt.title('Age of People')
plt.show()
```

Part 3: Access data from a dataset using functions like Numpy, Pandas

Now, for the above data create charts for all of them.

```python
import pandas as pd # Import the pandas library and alias it as 'pd'
import matplotlib.pyplot as plt # Import the matplotlib.pyplot library and alias it as 'plt'
import numpy as np

data1 = pd.read_csv('/content/HistoricalPrices.csv') # Use the correct function name
data1.head()
```

```python
# @title  Low vs  Close

from matplotlib import pyplot as plt
data1.plot(kind='scatter', x=' Low', y=' Close', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title  Open vs  High

from matplotlib import pyplot as plt
data1.plot(kind='scatter', x=' Open', y=' High', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title  Close

from matplotlib import pyplot as plt
data1[' Close'].plot(kind='hist', bins=20, title=' Close')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
